### Linear Regression ( Front thigh filter-1.0 )

###### x축-phase portrait 의 radius , y축-treadmil velocity

```python
# vel_05 리스트의 x와 y 값 생성
vel_05_x = [(value, 0.5) for value in vel_05]
# vel_08 리스트의 x와 y 값 생성
vel_08_x = [(value, 0.8) for value in vel_08]
vel_11_x = [(value, 1.1) for value in vel_11]
vel_14_x = [(value, 1.4) for value in vel_14]
vel_17_x = [(value, 1.7) for value in vel_17]
vel_20_x = [(value, 2.0) for value in vel_20]

vel_list = vel_05_x+vel_08_x+vel_11_x+vel_14_x+vel_17_x+vel_20_x
```

```python
# Linear regression 시각화
# x축이 phase portrait 의 radius, y축이 treadmil velocity

from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# 주어진 데이터에서 x와 y 값을 추출
X = np.array([pair[0] for pair in vel_list]).reshape(-1, 1)
y = np.array([pair[1] for pair in vel_list])

# 선형 회귀 모델 생성
model = LinearRegression()

# 데이터 학습
model.fit(X, y)

# 회귀 결과 확인
print("회귀 계수 (기울기):", model.coef_)
print("절편:", model.intercept_)

# 회귀선을 그리기 위한 x 값 생성
x_values = np.linspace(min(X), max(X), 100)
# 회귀선 예측
y_values = model.predict(x_values.reshape(-1, 1))

# 데이터와 회귀선 시각화
plt.scatter(X, y, color='blue', label='Data')  # 데이터 산점도
plt.plot(x_values, y_values, color='red', linewidth=2, label='Linear Regression')  # 회귀 선
plt.xlabel('X')  # x 축 라벨
plt.ylabel('Y')  # y 축 라벨
plt.title('Linear Regression')  # 그래프 제목
plt.legend()  # 범례 표시
plt.grid(True)  # 그리드 표시
plt.show()  # 그래프 출력
```
